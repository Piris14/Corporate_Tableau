```python
#cambiar nombres de columnas
df.rename(columns={'environmentScore':'Environment Score','GICS Sub-Industry':'Sub Industry','highestControversy':'Controversy Level','totalEsg':'ESG Total Score','socialScore':'Social Score','governanceScore':'Governance Score','percentile':'Percentile','ratingYear':'Rating Year', 'ratingMonth':'Rating Month','marketCap':'Market Cap','overallRisk':'Overall Risk'}, inplace=True)
```

```python
#Creo una columna con year y month
df['Rating Year'] = pd.to_numeric(df['Rating Year'], errors='coerce').astype('Int64')
df['Rating Month'] = pd.to_numeric(df['Rating Month'], errors='coerce').astype('Int64')

df['Rating Date'] = pd.to_datetime(
    {
        'year': df['Rating Year'],
        'month': df['Rating Month'],
        'day': 1
    },
    errors='coerce'
)
```

```python
numeric_cols = [
    'Environment Score', 'Social Score', 'Governance Score', 'ESG Total Score',
    'Controversy Level', 'Percentile', 'Market Cap', 'beta', 'Overall Risk'
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

```
